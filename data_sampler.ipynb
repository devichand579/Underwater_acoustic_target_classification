```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

from scipy.io import wavfile
import librosa
import librosa.display
import soundfile as sf
```

```python
def slice_and_save_audio(input_file, output_folder, sample_size_sec=1.0):
    # Load audio file
    y, sr = librosa.load(input_file, sr=52734)

    # Calculate number of samples per chunk and overlap
    samples_per_chunk = int(sample_size_sec * sr)

    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)
    x=0
    # Iterate through the audio signal, save each chunk as a separate file
    for i in range(0, len(y) - samples_per_chunk + 1, samples_per_chunk):
        x=x+1
        chunk = y[i:i + samples_per_chunk]
        output_file = os.path.join(output_folder, f"90__E__6H_{x}.wav")
        sf.write(output_file, chunk, sr)
```

```python
wav_path="/content/drive/MyDrive/final_dataset/class E/90__E__6H.wav"
sr=52734
data, fs = librosa.load(wav_path,sr=52734)
# fs is sampling frequency
# sampling frequency nothing but how may samples present for second.
print(f"Sampling frequency : {fs} and Wave : {data}")
```

```python
time = np.arange(0, len(data)) / sr
plt.plot(time, data)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
output_folder = "/content/drive/MyDrive/train_data_real/class_E"
sample_size_seconds = 0.5

slice_and_save_audio(wav_path, output_folder, sample_size_seconds)
```
